```python
operators_2022 = pd.read_csv('_datasets/_a/EIA923_2022_a.csv')
operators_2021 = pd.read_csv('_datasets/_a/EIA923_2021_a.csv')
operators_2020 = pd.read_csv('_datasets/_a/EIA923_2020_a.csv')
operators_2019 = pd.read_csv('_datasets/_a/EIA923_2019_a.csv')
operators_2018 = pd.read_csv('_datasets/_a/EIA923_2018_a.csv')
operators_2017 = pd.read_csv('_datasets/_a/EIA923_2017_a.csv')
```

```python
operators_2022
```

```python
operators_yoy = [operators_2017,
                 operators_2018,
                 operators_2019,
                 operators_2020,
                 operators_2021,
                 operators_2022]
```

```python
operators_yoy_df = pd.concat(operators_yoy)
```

```python
operators_yoy_df['YEAR'].value_counts()
```

```python
blank_years = operators_yoy_df[operators_yoy_df['YEAR'] == '.']
blank_years
```

```python
index_blank_years = operators_yoy_df[operators_yoy_df['YEAR'] == '.'].index
operators_yoy_df.drop(index_blank_years, inplace=True)
```

```python
operators_yoy_df['YEAR'].value_counts()
```

```python
operators_yoy_df.info()
```

```python
operators_yoy_df.rename(columns={'Reported\nPrime Mover': 'Prime Mover', 'YEAR': 'Year'})
```

```python
operators_yoy_df.to_csv('_datasets/operators_yoy.csv',index=False)
```
